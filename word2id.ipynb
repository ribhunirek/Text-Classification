```python
import pickle as pkl
import numpy as np
from collections import Counter
```

```python
uniq_words = []
vocab = []
data = []

with open('Common Unique Words', 'rb') as f:
    uniq_words = pkl.load(f)
    f.close()

with open('Vocab', 'rb') as f:
    vocab = pkl.load(f)
    f.close()

with open('Sentence Data', 'rb') as f:
    data = pkl.load(f)
    f.close()
```

```python
for w in uniq_words:
    if w not in vocab:
        vocab.append(w)
```

```python
vocab.append("Null")
```

```python
f = open('Vocab', 'wb')
pkl.dump(vocab, f)
f.close()
```

```python
v2 = len(vocab)
```

```python
#dictionary of word to idx in the Vocab list
word2id = {}

for idx, w in enumerate(vocab):
    word2id[w] = idx
```

```python
data
```

```python
for idx, row in enumerate(data):
    data[idx] = [word2id[w] if w in word2id else word2id["Null"] for w in row]
```

```python
# mean length of sentences in data is 20.26
# truncate sentences above 20 words and add padding for the ones less than 20
# padding is with word2id of the Null token
# although I'm going for lenth 10 as this is a learning project 

fix_len_data = []
fix_len = 10
pad = word2id['Null']

for sen in data:
    if len(sen) >= fix_len:
        fix_sen = sen[:fix_len]
    else:
        fix_sen = sen + [pad]* (fix_len - len(sen))
    fix_len_data.append(fix_sen)
```

```python
fix_len_data
```

```python
min(len(sen) for sen in fix_len_data)
```

```python
new_word_emb = list(np.random.randn(v2-v1, 50))
emb += new_word_emb
```
